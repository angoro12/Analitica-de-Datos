```python
import pandas as pd # Importar librería para leer datos en un dataframe
pd.set_option("display.max_columns", None)
import numpy as np # Importar librería numpy
import re # Importar librería para expresiones regulares  
import random # Librería para la generar números aleatorios 
```

Para tu conveniencia, la información ya se encuentra en un servidor IBM. Vamos a descargar los datos desde el servidor y leerlos en un dataframe llamado **récipes** (del inglés, recetas)


```python
recipes = pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DS0103EN/labs/data/recipes.csv")

print("Data read into dataframe!") # Toma alrededor de 30 segundos
```

Repetiremos los pasos de preprocesamiento implementados en el Laboratorio **De la Comprensión a la Preparación** con el fin de preparar el modelado de datos. Para mas detalles en como preparar los datos, revisa el Laboratorio **De la Comprensión a la Preparación**.


```python
# nombre de la columna que muestra la cocina
column_names = recipes.columns.values
column_names[0] = "cuisine"
recipes.columns = column_names

# convierte los nombres de la cocina en minúsculas 
recipes["cuisine"] = recipes["cuisine"].str.lower()

# hacer que los nombres de la cocina tengan consistencia
recipes.loc[recipes["cuisine"] == "austria", "cuisine"] = "austrian"
recipes.loc[recipes["cuisine"] == "belgium", "cuisine"] = "belgian"
recipes.loc[recipes["cuisine"] == "china", "cuisine"] = "chinese"
recipes.loc[recipes["cuisine"] == "canada", "cuisine"] = "canadian"
recipes.loc[recipes["cuisine"] == "netherlands", "cuisine"] = "dutch"
recipes.loc[recipes["cuisine"] == "france", "cuisine"] = "french"
recipes.loc[recipes["cuisine"] == "germany", "cuisine"] = "german"
recipes.loc[recipes["cuisine"] == "india", "cuisine"] = "indian"
recipes.loc[recipes["cuisine"] == "indonesia", "cuisine"] = "indonesian"
recipes.loc[recipes["cuisine"] == "iran", "cuisine"] = "iranian"
recipes.loc[recipes["cuisine"] == "italy", "cuisine"] = "italian"
recipes.loc[recipes["cuisine"] == "japan", "cuisine"] = "japanese"
recipes.loc[recipes["cuisine"] == "israel", "cuisine"] = "jewish"
recipes.loc[recipes["cuisine"] == "korea", "cuisine"] = "korean"
recipes.loc[recipes["cuisine"] == "lebanon", "cuisine"] = "lebanese"
recipes.loc[recipes["cuisine"] == "malaysia", "cuisine"] = "malaysian"
recipes.loc[recipes["cuisine"] == "mexico", "cuisine"] = "mexican"
recipes.loc[recipes["cuisine"] == "pakistan", "cuisine"] = "pakistani"
recipes.loc[recipes["cuisine"] == "philippines", "cuisine"] = "philippine"
recipes.loc[recipes["cuisine"] == "scandinavia", "cuisine"] = "scandinavian"
recipes.loc[recipes["cuisine"] == "spain", "cuisine"] = "spanish_portuguese"
recipes.loc[recipes["cuisine"] == "portugal", "cuisine"] = "spanish_portuguese"
recipes.loc[recipes["cuisine"] == "switzerland", "cuisine"] = "swiss"
recipes.loc[recipes["cuisine"] == "thailand", "cuisine"] = "thai"
recipes.loc[recipes["cuisine"] == "turkey", "cuisine"] = "turkish"
recipes.loc[recipes["cuisine"] == "vietnam", "cuisine"] = "vietnamese"
recipes.loc[recipes["cuisine"] == "uk-and-ireland", "cuisine"] = "uk-and-irish"
recipes.loc[recipes["cuisine"] == "irish", "cuisine"] = "uk-and-irish"


# borra los datos para cocinas con menos de 50 recetas
recipes_counts = recipes["cuisine"].value_counts()
cuisines_indices = recipes_counts > 50

cuisines_to_keep = list(np.array(recipes_counts.index.values)[np.array(cuisines_indices)])
recipes = recipes.loc[recipes["cuisine"].isin(cuisines_to_keep)]

# Convierte todos los 'Yes/ en 1 y los 'No' en 0
recipes = recipes.replace(to_replace="Yes", value=1)
recipes = recipes.replace(to_replace="No", value=0)
```

## \[bamboo_tree] Únicamente Cocina Asiática y de la India

En esta parte, vamos a crear un árbol de decisión para las recetas de ciertas cocinas asiáticas (coreana, japonesa, china, tailandesa) y de la India. La razón radica en que el árbol de decisión no se ejecuta apropiadamente cuando los datos están sesgados hacia una sola cocina, en este caso la norteamericana. Una opción es excluir la cocina americana de nuestro análisis o solamente construir arboles de decisión para distintos subconjuntos de datos. Escogeremos esto ultimo.


Construyamos arboles de decisión utilizando datos pertenecientes a las cocinas asiáticas e indu y después nombraremos a nuestro árbol de decisión como *bamboo_tree*.


```python
# Seleccionar el subconjunto de cocinas
asian_indian_recipes = recipes[recipes.cuisine.isin(["korean", "japanese", "chinese", "thai", "indian"])]
cuisines = asian_indian_recipes["cuisine"]
ingredients = asian_indian_recipes.iloc[:,1:]

bamboo_tree = tree.DecisionTreeClassifier(max_depth=3)
bamboo_tree.fit(ingredients, cuisines)

print("Decision tree model saved to bamboo_tree!")
```

Grafiquemos el árbol de decisión y a examinar como se ve.


```python
export_graphviz(bamboo_tree,
                feature_names=list(ingredients.columns.values),
                out_file="bamboo_tree.dot",
                class_names=np.unique(cuisines),
                filled=True,
                node_ids=True,
                special_characters=True,
                impurity=False,
                label="all",
                leaves_parallel=False)

with open("bamboo_tree.dot") as bamboo_tree_image:
    bamboo_tree_graph = bamboo_tree_image.read()
graphviz.Source(bamboo_tree_graph)
```

El arbol de decisión aprendió:

*   Si una receta contiene *cumin* (comino), *fish* (pescado) y **no** *yoghurt*, entonces es probable que sea una receta **Thai** (tailandesa).
*   Si una receta contiene *cumin* (comino) pero **no** *fish* (pescado) y **no** *soy_sauce* (salsa de soya) , entonces es probable que sea una receta **Indian** (indú).

# Model Evaluation <a id="4"></a>


<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DS0103EN/labs/images/lab4_fig5_flowchart_evaluation.png" width=500>


Para avaluar nuestro modelo de las cocinas asiática e hindú, dividiremos nuestro conjunto de datos un conjunto de entrenamiento y otro de pruebas. Construiremos el árbol de decisión usando el conjunto de entrenamiento. Después, probaremos el modelo con el conjunto de prueba y comparar la cocina que predice nuestro modelo con la actual.


Vamos a crear un nuevo dataframe utilizando solamente los datos pertenecientes a las cocinas asiática e hindú, y vamos a llamar **bamboo** al dataframe.


```python
bamboo = recipes[recipes.cuisine.isin(["korean", "japanese", "chinese", "thai", "indian"])]
```

Veamos cuantas recetas existen para cada cocina.


```python
bamboo["cuisine"].value_counts()
```

Borremos 30 recetas de cada cocina para usarlas como conjunto de prueba y llamaremos a este conjunto **bamboo_test**


```python
# establecer el tamaño de la muestra
sample_n = 30
```

Crear un dataframe que contenga 30 recetas de cada cocina seleccionadas aleatoriamente.


```python
# Tomar 30 recetas de cada cocina
random.seed(1234) # set random seed
bamboo_test = bamboo.groupby("cuisine", group_keys=False).apply(lambda x: x.sample(sample_n))

bamboo_test_ingredients = bamboo_test.iloc[:,1:] # ingredients
bamboo_test_cuisines = bamboo_test["cuisine"] # corresponding cuisines or labels
```

Revisar que haya 30 recetas de cada cocina


```python
# Revisar que tengamos 30 recetas de cada cocina
bamboo_test["cuisine"].value_counts()
```

A continuación, crearemos el conjunto de entrenamiento borrando el conjunto de prueba en **bamboo** y llamaremos a este **bamboo_train**.


```python
bamboo_test_index = bamboo.index.isin(bamboo_test.index)
bamboo_train = bamboo[~bamboo_test_index]

bamboo_train_ingredients = bamboo_train.iloc[:,1:] # ingredients
bamboo_train_cuisines = bamboo_train["cuisine"] # corresponding cuisines or labels
```

Observa que ahora hay *menos* de 30 recetas en cada cocina


```python
bamboo_train["cuisine"].value_counts()
```

Construyamos el árbol de decisión utilizando el conjunto de entrenamiento **bamboo_train** y llamémosle **bamboo_train_tree**.


```python
bamboo_train_tree = tree.DecisionTreeClassifier(max_depth=15)
bamboo_train_tree.fit(bamboo_train_ingredients, bamboo_train_cuisines)

print("Decision tree model saved to bamboo_train_tree!")
```

Vamos a graficar el árbol de decisión y a explorarlo.


```python
export_graphviz(bamboo_train_tree,
                feature_names=list(bamboo_train_ingredients.columns.values),
                out_file="bamboo_train_tree.dot",
                class_names=np.unique(bamboo_train_cuisines),
                filled=True,
                node_ids=True,
                special_characters=True,
                impurity=False,
                label="all",
                leaves_parallel=False)

with open("bamboo_train_tree.dot") as bamboo_train_tree_image:
    bamboo_train_tree_graph = bamboo_train_tree_image.read()
graphviz.Source(bamboo_train_tree_graph)
```

Ahora que definimos el árbol para que tenga mas profundidad, mas nodos de decisión se han generado.


#### Probemos nuestro modelo con los datos de prueba


```python
bamboo_pred_cuisines = bamboo_train_tree.predict(bamboo_test_ingredients)
```

Para saber que tan bueno es el árbol de decisión para determinar correctamente la cocina a la que pertenece cada receta, crearemos una matriz de confusión que contiene un buen resumen del numero de recetas de cada cocina clasificadas correctamente. Además se esclarece que cocinas están siendo confundidas con otras.


Sigamos adelante y creemos la matriz de confusión para saber que tan bien clasifica el árbol de decisión las recetas de **bamboo_test**


```python
test_cuisines = np.unique(bamboo_test_cuisines)
bamboo_confusion_matrix = confusion_matrix(bamboo_test_cuisines, bamboo_pred_cuisines, test_cuisines)
title = 'Bamboo Confusion Matrix'
cmap = plt.cm.Blues

plt.figure(figsize=(8, 6))
bamboo_confusion_matrix = (
    bamboo_confusion_matrix.astype('float') / bamboo_confusion_matrix.sum(axis=1)[:, np.newaxis]
    ) * 100

plt.imshow(bamboo_confusion_matrix, interpolation='nearest', cmap=cmap)
plt.title(title)
plt.colorbar()
tick_marks = np.arange(len(test_cuisines))
plt.xticks(tick_marks, test_cuisines)
plt.yticks(tick_marks, test_cuisines)

fmt = '.2f'
thresh = bamboo_confusion_matrix.max() / 2.
for i, j in itertools.product(range(bamboo_confusion_matrix.shape[0]), range(bamboo_confusion_matrix.shape[1])):
    plt.text(j, i, format(bamboo_confusion_matrix[i, j], fmt),
             horizontalalignment="center",
             color="white" if bamboo_confusion_matrix[i, j] > thresh else "black")

plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')

plt.show()
```

Después de ejecutar el código de arriba, deberás tener una matriz de confusión parecida a la siguiente:


<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DS0103EN/labs/images/lab4_fig6_confusion_matrix.png" width=500>


Las filas representan la cocina actual de este conjunto de datos y las columnas las que serán predichas. Cada fila debería sumar 100%. De acuerdo con esta matriz de confusión, haremos las siguientes observaciones:

*   Utilizando la primera fila en la matriz de confusión, 60% de las recetas **Chinese** (chinas) en **bamboo_test** fueron clasificadas correctamente por nuestro árbol de decisión mientras que el 37% de las mismas recetas fueron clasificadas erróneamente como **Korean** (coreanas) y un 3% como **Indian** (hindúes).

*   Haciendo uso de la fila Hindú, 77% de las recetas **Indian** en **bamboo_test** fueron clasificadas correctamente por el árbol de decisión pero el 3% y el 13% fueron clasificadas erróneamente como **Korean** y **Thai** respectivamente.
